```python
import pandas as pd

df = pd.read_csv("/content/KahootSummary.csv")
```

```python
df.head()
```

```python
import pandas as pd

df = df.drop(0)

new_header = df.iloc[0]
df = df[1:]
df.columns = new_header
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
#Questions with thev east 0? sort from least 0 to the q with most 0

import pandas as pd

zero_counts = {}


for column in df.columns:
  if 'Q' in column:
    zero_counts[column] = df[column].value_counts().get(0, 0)

sorted_questions = sorted(zero_counts.items(), key=lambda item: item[1])


for question, count in sorted_questions:
  print(f"{question}: {count} zeros")
```

```python

player_zero_counts = {}

# Iterate through rows (players) and count zeros
for index, row in df.iterrows():
  player_name = row['Player']
  zero_count = 0
  for column in df.columns:
    if 'Q' in column:
      if row[column] == '0':
        zero_count += 1
  player_zero_counts[player_name] = zero_count

# Sort players based on the number of zeros in descending order
sorted_players = sorted(player_zero_counts.items(), key=lambda item: item[1], reverse=True)


for player, count in sorted_players:
  print(f"{player}: {count} zeros")
```

```python

import matplotlib.pyplot as plt

question_difficulty = {}


for column in df.columns:
  if 'Q' in column:
    # Calculate the percentage of incorrect answers for each question
    incorrect_answers = df[df[column] == '0'][column].count()
    total_answers = df[column].count()
    difficulty_level = (incorrect_answers / total_answers) * 100
    question_difficulty[column] = difficulty_level


question_numbers = list(question_difficulty.keys())
difficulty_levels = list(question_difficulty.values())


plt.figure(figsize=(10, 6))
plt.bar(question_numbers, difficulty_levels)
plt.xlabel("Question Number")
plt.ylabel("Difficulty Level (%)")
plt.title("Difficulty Level for Each Question")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
